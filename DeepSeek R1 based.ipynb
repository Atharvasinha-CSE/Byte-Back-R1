```python
import gradio as gr
from transformers import pipeline
import re  # Import the regex library to filter text

# Load the model
bot = pipeline("text-generation", model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B", device_map="auto")

def chat(msg, history):
    # 1. System Prompt
    prompt = [{"role": "system", "content": "You are an elite, highly sarcastic Python developer."}]

    # 2. History
    for past_user_msg, past_bot_msg in history:
        prompt.append({"role": "user", "content": past_user_msg})
        prompt.append({"role": "assistant", "content": past_bot_msg})

    # 3. New Message
    prompt.append({"role": "user", "content": msg})

    # 4. The Brain (Increased max_new_tokens to 512 to allow room for "thinking")
    raw_output = bot(prompt, max_new_tokens=512)[0]['generated_text'][-1]['content']

    # 5. THE FIX: Remove the <think>...</think> block so the user only sees the final response
    clean_output = re.sub(r'<think>.*?</think>', '', raw_output, flags=re.DOTALL).strip()

    # Fallback just in case the model cuts off before finishing its thought
    if not clean_output:
        clean_output = raw_output

    return clean_output

# UI Setup
custom_theme = gr.Chatbot(avatar_images=["https://www.gravatar.com/avatar/?d=mp", "https://huggingface.co/datasets/huggingface/brand-assets/resolve/main/hf-logo.png"])
gr.ChatInterface(fn=chat, chatbot=custom_theme, title="🚀 My Industry AI", description="Powered by DeepSeek R1").launch(share=True)
```

```python
import gradio as gr
from transformers import pipeline
import re

# Load the model
bot = pipeline("text-generation", model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B", device_map="auto")

def chat(msg, history):
    # 1. STRICTER SYSTEM PROMPT: Tell it exactly how to behave and to avoid overthinking
    prompt = [{"role": "system", "content": "You are a sarcastic Python developer. Respond directly to the user in character. Do not explain your thought process or solve this prompt."}]

    # 2. History
    for past_user_msg, past_bot_msg in history:
        prompt.append({"role": "user", "content": past_user_msg})
        prompt.append({"role": "assistant", "content": past_bot_msg})

    # 3. New Message
    prompt.append({"role": "user", "content": msg})

    # 4. Generate
    raw_output = bot(prompt, max_new_tokens=512)[0]['generated_text'][-1]['content']

    # 5. ROBUST CLEANUP: Catches thoughts even if the model forgets the closing tag
    clean_output = re.sub(r'<think>.*?(</think>|$)', '', raw_output, flags=re.DOTALL|re.IGNORECASE).strip()

    return clean_output if clean_output else raw_output

# UI Setup
custom_theme = gr.Chatbot(avatar_images=["https://www.gravatar.com/avatar/?d=mp", "https://huggingface.co/datasets/huggingface/brand-assets/resolve/main/hf-logo.png"])
gr.ChatInterface(fn=chat, chatbot=custom_theme, title="🚀 My Industry AI", description="Powered by DeepSeek R1").launch(share=True)
```

```python
import gradio as gr
import re
from transformers import AutoModelForCausalLM, AutoTokenizer, TextIteratorStreamer
from threading import Thread

# 1. LOAD THE RAW MODEL (Graduating from 'pipeline')
model_name = "deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

def chat(msg, history):
    # 2. FORMAT THE PROMPT
    prompt = [{"role": "system", "content": "You are a sarcastic Python developer. Respond directly to the user in character. Do not explain your thought process."}]
    for past_user, past_bot in history:
        prompt.append({"role": "user", "content": past_user})
        prompt.append({"role": "assistant", "content": past_bot})
    prompt.append({"role": "user", "content": msg})

    # Convert the prompt into a format the raw model understands
    inputs = tokenizer.apply_chat_template(prompt, return_tensors="pt", add_generation_prompt=True).to(model.device)

    # 3. SET UP THE STREAMER
    streamer = TextIteratorStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)
    generation_kwargs = dict(inputs=inputs, streamer=streamer, max_new_tokens=512)

    # 4. START THE BACKGROUND THREAD
    thread = Thread(target=model.generate, kwargs=generation_kwargs)
    thread.start()

    # 5. CATCH AND YIELD WORDS LIVE
    partial_text = ""
    for new_token in streamer:
        partial_text += new_token

        # Dynamically hide the <think> block as it generates
        clean_output = re.sub(r'<think>.*?(</think>|$)', '', partial_text, flags=re.DOTALL|re.IGNORECASE).strip()

        # The 'yield' keyword is the magic that creates the typing animation
        if clean_output:
            yield clean_output
        else:
            yield "🤖 Thinking..."

# UI Setup
custom_theme = gr.Chatbot(avatar_images=["https://www.gravatar.com/avatar/?d=mp", "https://huggingface.co/datasets/huggingface/brand-assets/resolve/main/hf-logo.png"])
gr.ChatInterface(fn=chat, chatbot=custom_theme, title="🚀 My Industry AI", description="Powered by DeepSeek R1").launch(share=True)
```
